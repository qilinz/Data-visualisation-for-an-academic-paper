# Data visualisation for Van der Klink et al. (2001)
This report presents data visualisation based on paper Van der Klink et al. (2001), which employed meta-analysis to determine the effectiveness of occupational stress-reducing interventions.

```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from textwrap import wrap
%matplotlib inline
```

## Description of Studies
Number of studies in four intervention types: 5 studies evaluated an organization-focused intervention, 18 evaluated a cognitiveâ€“behavioral intervention, 17 evaluated a relaxation technique, and 8 evaluated a multimodal approach. 

```python
# Data frame: x-bars, y-height
height = [5,18,17,8]
bars = ('Organization-focused intervention', ' Cognitive â€“behavioral intervention', 'Relaxation technique', 'Multimodal approach')
# Resize the labels of bars @source: https://stackoverflow.com/questions/59466109/how-to-get-x-axis-labels-in-multiple-line-in-matplotlib
bars = ['\n'.join(wrap(x, 12)) for x in bars]

# Plot size
fig1 = plt.figure(figsize=(6,4))

# Create bars
plt.bar(bars, height)
 
# Create x-axis names
plt.xticks(bars)
 
# Label the data of bars
for i in range(len(height)):
    plt.annotate(str(height[i]), xy=(bars[i],height[i]), ha='center', va='bottom')

# Adjust y axis range
plt.ylim(0,20)

# Set title
plt.title(' Numbers of studies by the intervention categories')

# Show graph
plt.show()

# Save figure 1 (bbox_inches - make room for x axis labels https://stackoverflow.com/questions/6774086/why-is-my-xlabel-cut-off-in-my-matplotlib-plot
fig1.savefig('fig1.png', dpi = 300, bbox_inches = "tight")
```

## Effect sizes
A combined analysis of effect sizes yielded a significant effect size across all studies (d=0.34, 95% confidence interval)

Cohenâ€™s criteria (small effect: d<0.5; medium effect: 0.5<d<0.8; large effect: d>0.8)

 #### (1) Numbers of studies by effect sizes in Cohenâ€™s criteria: 
 17 significant (2-small, 4-medium, 11-large) and 31 non-significant

```python
# Data frame: x-bars, y-height
height = [2,4,11,31]
bars = ('Small effect','Medium effect','Large effect','Nonsignificance')

# Plot size
fig2 = plt.figure(figsize=(6,4))

# Create bars
plt.bar(bars, height)
 
# Create names on the x-axis
plt.xticks(bars)
 
# Label the data of bars
for i in range(len(height)):
    plt.annotate(str(height[i]), xy=(bars[i],height[i]), ha='center', va='bottom')

# Set title
plt.title('Numbers of studies by effect sizes in Cohenâ€™s criteria')

# Show graph
plt.show()

# Save figure 
fig2.savefig('fig2.png', dpi = 300)
```

#### (2) Effect sizes by types of interventions: 
The effect size of organizational-focused intervention is not significant while that of the rest three interventions are significant with p<0.05)

```python
# Dataframe
x= ('Organizational','Cognitiveâ€“behavioral','Relaxation','Multimodal')
y= [0.08,0.68,0.35,0.51]

# Confidence interval calculation
left = [-0.03, 0.54, 0.22, 0.33]
right = [0.19,0.82, 0.48, 0.69]
dy= np.array(right) - np.array(left)

# Plot
fig3 = plt.figure(figsize=(6,4))
plt.errorbar(x,y,yerr=dy,fmt='o',elinewidth=2,capsize=4)

# Set title
plt.title('Cohenâ€™s d and Confidence Intervals for the 4 Intervention Categories')

# Show
plt.show()

# Save
fig3.savefig('fig3.png', dpi = 300)
```

## Outcome Variables
#### (1) Two types of analysis are implemented: 

One is among all types of interventions and the other is among types of interventions that are focused on individuals (i.e. excluding organization-focused studies). 

Reason: effect sizes for outcome categories may be disproportionately affected by the small effect sizes found within organization-focused interventions as studies on organizational interventions usually involve significantly larger sample sizes than studies on individual interventions. 

```python
# Set the width of the bars
barWidth = 0.3
 
# Dataframe
organization = [0.17, 0.28, 0.30, 0.27, -0.03]
individual = [0.41, 0.48, 0.30, 0.42, -0.12]

# Plot size
fig4 = plt.figure(figsize=(6,4))

# The x position of bars
r1 = np.arange(len(organization))
r2 = [x + barWidth for x in r1]
 
# Create bars
plt.bar(r1, organization, width = barWidth, label='All types')
plt.bar(r2, individual, width = barWidth, label='Individuals only')

# Labels of x-axis
labels = ['Quality of work', 'Psychologic responses & resources', 'Physiology', 'Complaints', 'Absenteeism']
labels = ['\n'.join(wrap(x, 12)) for x in labels]

# Set title
plt.title("Cohen's d for results of all intervention types and interventions focusing on individuals")

# General layout
plt.xticks([r + barWidth for r in range(len(organization))], labels)
plt.ylabel("Cohen's d value")
plt.legend()

# Show graph
plt.show()

# Save
fig4.savefig('fig4.png', dpi = 300, bbox_inches = "tight")
```

With the exception of absenteeism, all effect sizes were significant at P<0.05 

#### (2) Interactions Between Intervention Types and Outcome Variables

```python
# Set the width of the bars
barWidth = 0.2
 
# Dataframe
cognitive = [0.48,0.65,0.11,0.52,-0.18]
relaxation = [0.29,0.26,0.31,0.31,-0.09]
multimodal = [0.59,0.22,0.36,0.48,0]
organizational = [0.05,0.14,0,0.05,0]

# Plot size
fig5 = plt.figure(figsize=(8,4))

# The x position of bars
r1 = np.arange(len(cognitive))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

# Create bars
plt.bar(r1, cognitive, width = barWidth, label='Cognitiveâ€“Behavioral')
plt.bar(r2, multimodal, width = barWidth, label='Multimodal')
plt.bar(r3, relaxation, width = barWidth, label='Relaxation')
plt.bar(r4, organizational, width = barWidth, label='Orgnizational')

# Labels of x-axis
labels = ['Quality of work', 'Psychologic responses & resources', 'Physiology', 'Complaints', 'Absenteeism']
labels = ['\n'.join(wrap(x, 12)) for x in labels]

# Set title
plt.title("Effect sizes in Cohen's d by interventions")

# General layout
plt.xticks([r + barWidth for r in range(len(cognitive))], labels)
plt.ylabel("Cohen's d value")
plt.legend()

# Show graph
plt.show()

# Save
fig5.savefig('fig5.png', dpi = 300, bbox_inches = "tight")
```

(a) Quality of work: multimodal intervention is most effective.

(b) Psychologic responses and resources: cognitive-behavioral intervention is most effective.

(c) Psychophysiologic outcomes (Physiology in figure): relaxation intervention  and multimodal intervention are more effective.

(d) Reducing employeesâ€™ stress-related complaints (Complaints in figure): cognitive-behavioral intervention and multimodal intervention are more effective.

(e) Absenteeism: no intervention is effective.

(f) The organizational intervention is only significant in psychologic responses and resources category.
